```python
class Coffeshop:
    
    def __init__(self,name,menu,orders):
        
        self.name =name
        self.menu =menu 
        self.orders=orders 
        
    
    def add_order(self,item):
        c=False
        
        for i in self.menu:
            for key,value in i.items():
                if value==item:
                    c=True
                    self.orders.append(item)
                    break
                else:
                    c=False
        if c==True:
            print("Order is successfully placed ")
            return self.orders
            
        elif c==False:
            return "Item is currently unavailable"
        
            
    def fulfil_order(self):
        orders_list =self.orders
        for i in range(0,len(orders_list)):
            print(" The {} is ready".format(orders_list[i]))
        
        else:
            print("order is fulfilled")
            
        
        
            
    def list_order(self):
        
        if len(self.orders) !=0:
            print(" Your list of order is {}".format(self.orders))
         
        else:
            print("[]")
            
    def due_amount(self):
        sum=0 
        for item in self.orders:
            for i in self.menu:
                if i['name']==item:
                    sum=sum+i["price"]
                    
        print(" Total due amount for your order is {}".format(sum))
        
    def cheapest_item(self):
        list_price =[]
        for i in self.menu:
            list_price.append(i['price'])
        
        ele =sorted(list_price)[0]
        
        for i in self.menu:
            if i['price'] ==ele:
                print(" Chepeast item sir is {}".format( i['name']))
                
    
    def drinks_only(self):
        list_drinks= []
        for i in self.menu:
            if i['type'] =="Drinks":
                list_drinks.append(i['name'])
                
        print(" the items with the drinks only are {}".format( list_drinks))
        
    def food_only(self):
        list_food= []
        for i in self.menu:
            if i['type'] =="Food":
                list_food.append(i['name'])
                
        print(" items with only food type are {}".format(list_food))
                
                
c=Coffeshop("Restro" ,[{"name" :"Chicken" ,"type":"Food" ,"price" :12500} ,{"name": "Prawn" ,"type":"Food","price":15000} ,{"name": "Tequilla" ,"type":"Drinks","price":15000} ] ,[])
c.add_order("Prawn")
c.add_order("Chicken")
c.fulfil_order()
c.list_order()
c.due_amount()
c.cheapest_item()
c.drinks_only()
c.food_only()
```
